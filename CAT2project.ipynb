```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
# Load the voter registration dataset and each social media dataset into separate pandas dataframes
voter1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/voter-matches.csv')
facebook = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/pseudo_facebook.csv')
twitter = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/twitter_dataset.csv')
instagram = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset.csv')
# Merge the social media dataframes into a single dataframe
social_media_df = pd.merge(facebook, twitter, on='userid')
social_media_df = pd.merge(social_media_df, instagram, on='userid')
# Preprocess the data by removing any unnecessary columns and handling missing values
social_media_df = social_media_df.drop(['Account_start_time','Username','Text', 'Sex', 'Twitter_username', 'Name','Instagram_username', 'Timestamp', 'Account_start_time'], axis=1)
social_media_df = social_media_df.fillna(0)
# Convert categorical variables to numerical using one-hot encoding
categorical_cols = ['state','party']
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_cols = encoder.fit_transform(voter1[categorical_cols])
feature_names = encoder.get_feature_names_out(categorical_cols)
encoded_df = pd.DataFrame(encoded_cols, columns=feature_names)
voter1_encoded = pd.concat([voter1.drop(categorical_cols, axis=1), encoded_df], axis=1)
# Merge the voter registration dataframe with the social media dataframe
merged_df = pd.merge(voter1_encoded, social_media_df, on='userid')
# Split the data into training and testing sets
X = merged_df.drop(['userid', 'voterid','gender'], axis=1)
y = merged_df['Age']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Train a linear regression model
linear_model = LinearRegression()
linear_model.fit( X_train, y_train)
# Make predictions on the test set using linear regression
y_pred_linear = linear_model.predict(X_test)
# Evaluate the linear regression model using mean squared error and R-squared score
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
mse_linear, r2_linear
```
